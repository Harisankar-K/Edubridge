# Data Visualisation

### goals vs players using pie chart 

```python
plt.pie(euro['Goals'],labels=euro['Player'],startangle=90,shadow=True)
plt.legend(loc='center',title='report')
plt.show()
```

### goals vs assists comparison using barplot

```python
plt.figure(figsize=(12,8))
plt.title('goals vs assist comparison')
sns.barplot(x='Goals',y='Assists',data=euro,hue='Position',palette='tab10')
plt.xlabel('Goals',fontsize=15)
plt.show()
```

### euro vs assists using bar graph

```python
plt.figure(figsize=(10,7))
plt.bar(euro['Goals'],euro['Assists'],color='b')
plt.show()
```

### goals vs  left footed goals by bargraph 

```python
plt.bar(euro['RFG'],euro['Goals'])
plt.title('right footed goals',fontsize=15,color='red')
plt.xlabel('RFG',fontsize=15)
plt.ylabel('Goals',fontsize=15)
plt.grid(True)
plt.show()
```

### plot of goals using countplot

```python
sns.countplot(y='Goals',data=euro,palette='tab10')
plt.show()
```

### pie chart of  position

```python
plt.title('postion',fontsize=25)
euro['Position'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,explode=(0.5,0,0))
plt.show()
```

### position vs assists / goals using bargraph

```python
euro.groupby('Position')[['Assists','Goals']].sum().plot.bar(color=['blue','red'],figsize=(8,6))
plt.ylabel('assists/Goals')
plt.show()
```

### position vs minutes played using stripplot

```python
sns.set_style('darkgrid')
sns.stripplot(x='Position',y='mp',jitter=0.40,dodge=True,data=euro)
plt.show()
```

### minutes played vs country using relational plot 

```python
plt.figure(figsize=(16,18))
sns.relplot(x='mp',y='Country',hue='Position',kind='line',data=euro)
plt.show()
```

### goals vs matches played used line plot

```python
sns.lineplot(x='Goals',y='Match played',hue='Position',data=euro)
plt.show()
```

### matches played  vs minutes played using catplot

```python
sns.catplot(x="Match played",y="mp",hue='Position',kind="boxen",data=euro)
plt.show()
```

### assists vs goals using kdeplot

```python
sns.kdeplot(euro['Assists'],euro['Goals'],color='grey',vertical=True,shade=True)
plt.show()
```

### pair plot europlayers in england

```python
sns.pairplot(euro_england)
plt.show()
```

### left footed goals vs goals using violin plot

```python
sns.violinplot(x='LFG',y='Goals',hue='Position',data=euro,size=50)
plt.legend(loc='upper right')
plt.show()
```

### spain players using violin plot

```python
euro=euro[euro.Country=='Spain']
sns.violinplot(x=euro.Country,y=euro.mp,color='yellow')
plt.show()
```

### goals vs players using swarmplot

```python
sns.swarmplot(x='Goals',y='Player',hue='Position',dodge=True,data=euro,size=10)
plt.show()
```

### players vs minutes played of spain players using violin and swarm plot

```python
euro=euro[euro.Country=='Spain']
sns.violinplot(x=euro.Player,y=euro.mp,scale='width',inner=None,color='xkcd:forest green'
              )
sns.swarmplot(x=euro.Player,y=euro.mp,color='yellow',size=10
             )
plt.show()
```

### country vs goals using boxplot
