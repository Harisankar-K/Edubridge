# Data Visualisation

###  petrol price highest using bargraph: 

it shows that highest petrol price was noted in the october of 2018

```python
plt.figure(figsize=(10,15))
plt.bar(petrol['Highest'],petrol['month']) 
plt.title('petrol highest price',fontsize=15,color='red') 
plt.xlabel('highest',fontsize=15) 
plt.ylabel('month',fontsize=15) 
plt.grid(False)
plt.show()
```

### horizontal bar graph:
in this we can see that the price of petrol in each months and lowest petrol price was noted in the month jan of 2019

```python
plt.figure(figsize=(10,15))
plt.barh(petrol['month'],petrol['Lowest'],color="hotpink") 
plt.title('petrol lowest price',fontsize=15,color='hotpink') 
plt.xlabel('lowest',fontsize=15) 
plt.ylabel('month',fontsize=15) 
plt.grid(False)
plt.show()
```

###  highest petrolprice using count plot

```python
plt.figure(figsize=(5,10))
sns.countplot(y='Highest',data=petrol,palette='tab10')
plt.show()
```

### splitting the dataset highest / lowest on the basis of month
in this we can easily analyse that the price if petrol lowest and highest of each month respectively.lowest was noted by thw colour red  and  highest was noted by the colour violet, and also we can analyse that the highest price for petrol  was noted in the month october of 2018 and lowet was noted in the month january 2019. 

```python
petrol.groupby('month')[['Highest','Lowest']].sum().plot.bar(color=['blue','red'],figsize=(8,6))
plt.ylabel("highest/Lowset")
plt.show()
```

### strip plot:
in this we can analys that the highest petrol rate was on october of 2018 

```python
plt.figure(figsize=(10,15))
sns.set_style('darkgrid')
sns.stripplot(x='Highest',y='month',data=petrol)

plt.show()
```

### line plot:
in this we can analys that the highest petrol price was marked on the october of 2018

```python
plt.figure(figsize=(10,15))
sns.lineplot(x='Highest',y='month',data=petrol)
plt.show()
```

### scater plot 
we can analyse the relationship between the highest and month

```python
plt.figure(figsize=(10,15))
sns.scatterplot(x='Highest',y='month',data=petrol,palette="tab10")
plt.show()
```

### boxplot
in this we can see that the Highest petrol price is noted in the month of october 2018 using boxplot

```python
plt.figure(figsize=(10,15))
sns.boxplot(x='Highest',y='month',data=petrol,width=0.5)
plt.show()
```

###  pair plot:

###  specify the split size manually by specifying the parameter of the test size and random number generator

### linear regression analysis

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x_train,y_train)
print('train test complete')
```

```python
y_pred=l.predict(x_test)
```

```python
petrol=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
petrol
```

```python
from sklearn.metrics import mean_absolute_error
print('Mean Absolute Error',mean_absolute_error(y_test,y_pred))

```

```python
sns.regplot(x=y_test,y=y_pred)
plt.title('Regession Plot',size=15)
plt.ylabel('predicted',size=12)
plt.xlabel('actual',size=12)
plt.show()
```
